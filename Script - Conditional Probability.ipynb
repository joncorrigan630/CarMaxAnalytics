```python
import numpy as np
import pandas as pd 
```

```python
import_file_path = "Vehicle_Purchase_Data.csv"
import_data = pd.read_csv(import_file_path, index_col=0)
dataset = import_data.copy()
dataset 
```

```python
purchases_df = dataset.loc[:, ["previous_purchase", "subsequent_purchases"]]
purchases_df
```

```python
prev_purchase_proportion = purchases_df["previous_purchase"].value_counts(normalize=True, sort=True).reset_index()
prev_purchase_proportion
```

```python
prev_purchase_count = purchases_df["previous_purchase"].value_counts(sort=True).reset_index()
prev_purchase_count
```

```python
prev_purchase_merged = prev_purchase_count.merge(prev_purchase_proportion, on='index')
prev_purchase_merged.rename(columns = {'index':'prev_purchase_categories', 'previous_purchase_x':'previous_purchase_count', 'previous_purchase_y':'previous_purchase_proportion'}, inplace=True)
prev_purchase_merged
```

```python
subseq_purchase_proportion = purchases_df["subsequent_purchases"].value_counts(normalize=True, sort=True).reset_index()
subseq_purchase_proportion 
```

```python
subseq_purchase_count = purchases_df["subsequent_purchases"].value_counts(sort=True).reset_index()
subseq_purchase_count
```

```python
subseq_purchase_merged = subseq_purchase_count.merge(subseq_purchase_proportion, on='index') 
subseq_purchase_merged.rename(columns = {'index': 'subsequent_purchase_categories', 'subsequent_purchases_x':'subsequent_purchase_count', 'subsequent_purchases_y':'subsequent_purchase_proportion'}, inplace=True)
subseq_purchase_merged 
```

Pr(A) = Previous Purchase - TRUE 

Pr(B) = Subsequent Purchases - TRUE

Pr(A^c) = Previous Purchase - FALSE

Pr(B^c) = Subsequent Purchases - FALSE

If Previous Purchase is FALSE, then it is the first purchase

If Previous Purchase is TRUE, then it is not the first purchase

If Subsequent Purchases is FALSE, then 0 subsequent purchases have been made 

If Subsequent Purchases is TRUE, then 1,2,3...10 subsequent purchases have been made

If Previous Purchase is FALSE & Subsequent Purchases is FALSE, then 0 subsequent purchases have been made since the first purchase

If Previous Purchase is FALSE & Subsequent Purchases is TRUE, then >= 1 subsequent purchases have been made since the first purchase

If Previous Purchase is TRUE & Subsequent Purchases is TRUE, then this is not the first purchase and subsequent purchases have been made

If Previous Purchase is TRUE & Subsequent Purchases is is FALSE, then this is not the first purchase and no subsequent purchases have been made

```python
prob_a = round(prev_purchase_merged["previous_purchase_count"][1] / len(purchases_df), 3)
prob_not_a = round(prev_purchase_merged["previous_purchase_count"][0] / len(purchases_df), 3)
prob_b = (subseq_purchase_merged["subsequent_purchase_count"].sum()) - (subseq_purchase_merged["subsequent_purchase_count"][0]) 
prob_b = round(prob_b  / len(dataset), 3)  
prob_not_b = round(1 - prob_b, 3)  

prob_a_intersection_b = round(prob_a * prob_b, 3) 
prob_not_a_intersection_not_b = round(prob_not_a * prob_not_b, 3) 
prob_a_intersection_not_b = round(prob_a * prob_not_b, 3)
prob_not_a_intersection_b = round(prob_not_a * prob_b, 3) 

prob_a_union_b = round((prob_a + prob_b) - (prob_a * prob_b), 3)
prob_not_a_union_not_b = round((prob_not_a + prob_not_b) - (prob_not_a * prob_not_b), 3) 
prob_a_union_not_b = round((prob_a + prob_not_b) - (prob_a * prob_not_b), 3) 
prob_not_a_union_b = round((prob_not_a + prob_b) - (prob_not_a * prob_b), 3) 

prob_a_given_b = round(prob_a_intersection_b / prob_b, 3)
prob_not_a_given_not_b = round(prob_not_a_intersection_not_b / prob_not_b, 3) 
prob_a_given_not_b = round(prob_a_intersection_not_b / prob_not_b, 3) 
prob_not_a_given_b = round(prob_not_a_intersection_b  / prob_b, 3) 

prob_b_given_a = round(prob_a_given_b * prob_b / prob_a_given_b * prob_b + prob_a_given_not_b * prob_not_b, 3) 
prob_b_given_not_a = round(prob_not_a_given_b * prob_b / prob_not_a_given_b * prob_b + prob_not_a_given_not_b * prob_not_b, 3) 
prob_not_b_given_a = round(prob_a_given_not_b * prob_not_b / prob_a_given_not_b * prob_not_b + prob_a_given_b * prob_b, 3)  
prob_not_b_given_not_a = round(prob_not_a_given_not_b * prob_not_b / prob_not_a_given_not_b * prob_not_b + prob_not_a_given_b * prob_b, 3)  
```

```python
print("Probability of A: Pr(A) = {}".format(prob_a))
print("Probability of B: Pr(B) = {}".format(prob_b)) 
print("Probability of A Complement: Pr(A^c) = {}".format(prob_not_a)) 
print("Probability of B Complement: Pr(B^c) = {}".format(prob_not_b)) 

print("") 

print("Probability of A Intersection B: Pr(A n B) = {}".format(prob_a_intersection_b)) 
print("Probability of A Complement Intersection B Complement: Pr(A^c n B^c) = {}".format(prob_not_a_intersection_not_b))
print("Probability of A Intersection B Complement: Pr(A n B^c) = {}".format(prob_a_intersection_not_b))  
print("Probability of A Complement Intersection B: Pr(A^c n B) = {}".format(prob_not_a_intersection_b)) 

print("")

print("Probability of A Union B: Pr(A u B) = {}".format(prob_a_union_b)) 
print("Probability of A Complement Union B Complement: Pr(A^c u B^c) = {}".format(prob_not_a_union_not_b)) 
print("Probability of A Union B Complement: Pr(A u B^c) = {}".format(prob_a_union_not_b)) 
print("Probability of A Complement Union B: Pr(A^c u B) = {}".format(prob_not_a_union_b))  

print("")

print("Probability of A given B: Pr(A|B) = {}".format(prob_a_given_b))  
print("Probability of A Complement given B Complement: Pr(A^c|B^c) = {}".format(prob_not_a_given_not_b))  
print("Probability of A given B Complement: Pr(A|B^c) = {}".format(prob_a_given_not_b)) 
print("Probability of A Complement given B: Pr(A^c|B) = {}".format(prob_not_a_given_b)) 

print("")

print("Probability of B given A: Pr(B|A) = {}".format(prob_b_given_a)) 
print("Probability of B given A Complement: Pr(B|A^c) = {}".format(prob_b_given_not_a))  
print("Probability of B Complement given A: Pr(B^c|A) = {}".format(prob_not_b_given_a))  
print("Probability of B Complement given A Complement: Pr(B^c|A^c) = {}".format(prob_not_b_given_not_a)) 
```
